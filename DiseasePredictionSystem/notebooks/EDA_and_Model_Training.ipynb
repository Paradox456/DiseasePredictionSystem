```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/zihan/DiseasePredictionSystem/data/Testing.csv")
print(df)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
disease_counts = df['prognosis'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(disease_counts, labels=disease_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Disease Distribution')
plt.show()
```

```python
df.hist(bins=50, figsize=(50,100))
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x=df["itching"])
plt.title("Box Plot of Numerical Feature")
plt.show()
```

```python
sns.countplot(x='prognosis', data=df)
plt.xticks(rotation=90)
plt.title("Distribution of Target Variable")
plt.show()
```

```python
sns.countplot(x="itching", hue="prognosis", data=df)
plt.title("Categorical Feature vs Target Variable")
plt.show()
```
